```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
x = []
y = []
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
for i in range(-6, 7):
    y.append(sigmoid(i))
    x.append(i)
print(x,y)

rcParams['figure.figsize'] = 3, 3
rcParams["legend.loc"] = 'upper left'
rcParams['axes.labelsize'] = 10
rcParams['axes.titlesize'] = 5
rcParams["font.size"] = 10
plt.axvline(0, color='black')
plt.axhline(0, color='black')
plt.plot(x,y)
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

```python
x = []
y = []
def ReLU(z):
    return z* (z > 0)
for i in range(-6, 7):
    y.append(ReLU(i))
    x.append(i)
print(x,y)
plt.axvline(0, color='black')
plt.axhline(0, color='black')
plt.plot(x,y)
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

```python
x = []
y = []
def tanh(z):
    return (2/(1 + np.exp(-2*z))-1)

for i in range(-6, 7):
    y.append(tanh(i))
    x.append(i)
print(x,y)
plt.axvline(0, color='black')
plt.axhline(0, color='black')
plt.plot(x,y)
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```
